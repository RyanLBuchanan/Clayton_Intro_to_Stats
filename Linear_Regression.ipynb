```python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(19680801)


N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
Churn = pd.read_csv('Data/churn_clean.csv')
Churn['intercept'] = 1
lm_Churn = sm.OLS(Churn['MonthlyCharge'], Churn[['Income', 'intercept']]).fit()
residuals = Churn['MonthlyCharge'] - lm_Churn.predict(Churn[['Income', 'intercept']])
sns.scatterplot(x=Churn['MonthlyCharge'], y=residuals, color='red')
plt.show();
```
